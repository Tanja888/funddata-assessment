_Adapt the code to generate a 1000 character long string with weights based on the previous two characters._  

```python
# Make HTTP requests for internet reqources.
import urllib.request

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]

# Get a sample paragraph
paragraph = ' '.join(book[58:63]) #join method to turn a paragraph list into a string

# Show the paragraph.
print(paragraph)
```

```python
# Let's lower-case it.
alice = paragraph.lower()

# All letters and a space.
chars = 'abcdefghijklmnopqrstuvwxyz '

# And strip anything that is not a letter or space.
alice = ''.join([c for c in alice if c in chars]) #have to look more into list comprehension

# Show the paragraph now.
print(alice)
```

```python
# For doing (pseudo-)random things in Python.
import random

# Print a random character, for example.
print(random.choice(chars))
```

```python
# Get the length of alice.
N = len(alice)

# Generate N random characters from chars.
gener = random.choices(chars, k=N) #k is the lentgh of the output

# Join them together in a string.
gener = ''.join(gener)

# Print.
print(gener)
```

```python
# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

```python
# Generate a string using those weights.
wgenr = random.choices(chars, weights=weights, k=N)

# Join them together in a string.
wgenr = ''.join(wgenr)

# Print.
print(wgenr)
```

```python
# First created the weights based on the previous character
# Added one more outer dictionary e to get the weights based on the previous two characters
twoghts = {e: {c: {d: sbook.count(c + d + e) for d in chars} for c in chars} for e in chars}

# Show the weights.
twoghts
```

```python
# Loop through our character set.
for i in range(len(chars)):
    # Print the character and how many times it appears in Alice in Wonderland.
    print(f'{chars[i]}: {weights[i]}')
```

#### Exercise 2 <br>
Explain why the log of zero is undefined.

Following the rules for calculating logarithms:  
we cannot have the base 0 in the logarithmic function and we cannot have 0 as an argument.

First, we can turn logarithmic into an exponential function: 
    
$log_{b}$a = x  
a = $b^{x}$

If we try to use 0 as the base we get the false result:  
$log_{0}$2 = x     
2 $\neq$ $0^{x}$
    
If we use 0 as an argument we get this function:  
    $log_{b}0$ = x  
    
Since we know the base cannot be 0 this can never be true, thus the log of 0 is undefined.   
    0=$b^{x}$
